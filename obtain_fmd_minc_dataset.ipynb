# Obtain Minc dataset

For strange reasons, colab does not extract/unzip the MINC dataset fully sometimes, and sometimes colab data does not sync with gdrive even when drive is mounted

To ensure the complete data of the ten required classes from MINC are unzipped, saved and synced with gdrive, do the following: 

1. Download MINC to your local computer and unzip it locally

2. On your local computer, zip **each** of the ten image folders (fabric, foliage, glass, leather, metal, paper, plastic, stone, water, and wood) in ```MINC/minc-2500/images``` **individually**

3. Run the cell right below to create ```MINC/minc-2500/images``` in the CZ4042_project folder on GDrive

4. Upload the ten zipped folders to ```MINC/minc-2500/images``` on GDrive

5. Run the ten cells below to unzip the ten zipped files that you uploaded to gdrive. Note that each time when ```drive.flush_and_unmount()``` is called, the colab content will be explicitly sync with gdrive. Although this ensures the files unzipped via colab are guaranteed to be synced with your gdrive, you will need to restart the colab notebook before you can run the next cell to unzip the next zipped file. This means you need to manually restart and run the notebook for each of the ten unzippings.

6. Once verified each unzipped folder contain 2500 images and can be found on gdrive, remove the ten zipped files from gdrive

````python
if not os.path.exists('MINC'):
    os.makedirs('MINC')
    os.chdir('MINC')
    os.makedirs('minc-2500')
    os.chdir('minc-2500')
    os.makedirs('images')
    os.chdir('images')
else:
  os.chdir('MINC/minc-2500/images')
````
